### Analysis 1:

```python
# 1. Create a histogram of the number of observations per species
# Create a dataframe with the number of observations per species
species = artis_observations['scientificName'].value_counts()
species = species.to_frame()
species = species.reset_index()
species.columns = ['species', 'number_of_observations']

# Create a histogram of the number of observations per species
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="species", y="number_of_observations", data=species)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.show()
```

### Analysis 2: 

```python
# Create a dataframe with the number of observations per year
rattus_norvegicus = artis_observations[artis_observations['scientificName'] == 'Rattus norvegicus']
```

### Analysis n:

```python
# visualise when during the day the rattus norvegicus was observed
# 1. Create a histogram of the number of observations per hour
# Create a dataframe with the number of observations per hour
rattus_norvegicus['hour'] = rattus_norvegicus['timestamp'].str.split('T').str[1].str.split(':').str[0]
rattus_norvegicus['hour'] = rattus_norvegicus['hour'].astype(int)
rattus_norvegicus_hour = rattus_norvegicus['hour'].value_counts()
rattus_norvegicus_hour = rattus_norvegicus_hour.to_frame()
rattus_norvegicus_hour = rattus_norvegicus_hour.reset_index()
rattus_norvegicus_hour.columns = ['hour', 'number_of_observations']

# Create a histogram of the number of observations per hour
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="hour", y="number_of_observations", data=rattus_norvegicus_hour)
plt.show()
```
